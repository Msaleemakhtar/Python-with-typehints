```python
class StudentData():
    def __init__(self, roll_no:int, name:str, age:int)->None:
        if age < 18 or age > 60:
            raise StudentAgeError("You are not eligible")
        self.roll_no = roll_no
        self.name = name
        self.age = age

class StudentAgeError(Exception):
    pass

student1 = StudentData(2, "saleem", 42)
print(student1.name, student1.age, student1.roll_no)
```

```python
with open("./abc.txt") as data:
    print(data.readline())
    print(type(data))
    print(data.readline())# read line one at a time


print("pakistan")
```

```python
with open("./abc.txt") as data:
   
    print(data.readlines())# all lines 


print("pakistan")
```

```python
with open("./abc.txt", "r") as data:
    print(data.readline())
    print(data.write("pakistan"))
  


print("pakistan")
```

```python
from pydub import AudioSegment

audio = AudioSegment.from_file("./audio.wav")
print("Channels:", audio.channels)
print("Sample Width:", audio.sample_width)
print("Frame Rate:", audio.frame_rate)
print("Frame Width:", audio.frame_width)
print("Length (ms):", len(audio))
print("Frame Count:", audio.frame_count())
```

```python
import pygame

# Initialize pygame
pygame.init()

# Set the file path for your audio file
#audio_file = r"C:\Users\saram\OneDrive\Desktop\python\error_file handling\audio.wav"

# Load the audio file
pygame.mixer.music.load("./audio.wav")

# Play the audio
pygame.mixer.music.play()

# Wait for the audio to finish playing
while pygame.mixer.music.get_busy():
    pygame.time.delay(100)

# Quit pygame
pygame.quit()
```
