# Pandas core components--two basic types 
   * series

```python
# # list can be used for creating series

s1:pd.Series= pd.Series([1,3,4,5,6,7])
print(s1)
```

```python
# Tuple can be used for creating series

s1:pd.Series= pd.Series((1,3,4,5,6,7))
print(s1)
```

```python
# Dictionary can  be used for creating series

s1:pd.Series= pd.Series({"name":"saleem", "fname": "GM"})
print(s1[1])
```

```python
# # set cannot be used for creating series

s1:pd.Series= pd.Series({1,3,4,5,6,7})# type is unordered
print(s1)
```

```python
values :list[int]  = [1,2,3,4,5]
value :list[str]  = ["a", "c", "d", "e", "f"]
match :pd.Series = pd.Series(values, index=value)
print(match)
```

```python
values :list[int]  = [100,200,300,400,500]
# multi indexing
value :list[list[str]]  = [["a1", "a1", "b1", "b1", "b1"],
                           ["a", "c", "d", "e", "f"]]
match :pd.Series = pd.Series(values, index=value, name = "Student data")
print(match)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Set a seed for reproducibility
np.random.seed(0)

# Define the number of transactions
num_transactions = 1000

# Generate random dates within the current year
current_year = datetime.now().year
print(current_year)
start_date = datetime(current_year, 1, 1)
print(start_date)
end_date = datetime.now()
print(end_date)
print((end_date - start_date).days)

random_dates = [start_date + timedelta(days=np.random.randint(0, (end_date - start_date).days)) 
                for _ in range(num_transactions)]
print(len(random_dates))

# Generate a list of random fee amounts
random_fees = np.random.randint(low=1, high=5000, size=num_transactions)

# Create the DataFrame
fee_transactions = pd.DataFrame({
    'date': random_dates,
    'fee': random_fees
})

print(fee_transactions)
print("------------")

# Sort by date for readability
fee_transactions.sort_values('date', inplace=True)
print(fee_transactions)
print("------------")
# Reset index after sorting
fee_transactions.reset_index(drop=True, inplace=True)

print(fee_transactions)
```
